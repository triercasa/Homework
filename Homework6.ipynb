```python
import numpy as pd
import pandas as pd
df=pd.read_excel('sampledatafood.xlsx', index_col=0)
```

```python
print(df)
```

```python
print(df.head())  # Display the first 5 rows
print(df.describe())  # Display all contents (summary statistics)
```

```python
#Display all the product names,  and unit price separated by commas.
print("Product Name, Unit Price")
for index, row in df.iterrows():
    print(f"{row['Product']}, {row['UnitPrice']:.2f}")
```

```python
#import to use basic plotting
import matplotlib.pyplot as plt

# Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['UnitPrice'], color='salmon') #because I like the color salmon
plt.xlabel('Product')
plt.ylabel('UnitPrice')
plt.xticks(rotation=45, ha='right')
plt.show()
```
